Stwórz dane z rozkładu normalnego o średniej 3 i odchyleniu standardowym 1 dla 100 oraz 10 000
Na podstawie tych sztucznie stworzonych zbiorów oblicz statystyki poznane w module statystki opisowej.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Ustawienie ziarna generatora liczb losowych dla powtarzalności wyników
np.random.seed(0)

# Generowanie danych:
# 100 obserwacji
dane_100 = np.random.normal(loc=3, scale=1, size=100)
# 10 000 obserwacji
dane_10000 = np.random.normal(loc=3, scale=1, size=10000)
```

Plotowanie histogramu ze zbiorem 1

```python
counts, bins, patches = plt.hist(dane_100, bins=20, density=False, color='skyblue',edgecolor='black', alpha=0.7, histtype='bar', rwidth=0.9, label='Dane')
plt.xlabel("Wartosc")
plt.ylabel("Ilosc probek na 100")
for count, bin_edge in zip(counts, bins):
    if count > 0:
        plt.text(bin_edge + (bins[1] - bins[0]) / 2, count, f'{int(count)}', 
                 ha='center', va='bottom', fontsize=8)
mean_100 = np.round(np.mean(dane_100))
plt.axvline(x = mean_100, color='red', label = 'Średnia')
median_100 = np.median(dane_100)
plt.axvline(x = median_100, color='green', label = 'Mediana')
standard_deviation_100 = np.std(dane_100, ddof=1)
print(f"Standard deviation {standard_deviation_100}")
plt.axvline(mean_100 + standard_deviation_100, color='red', linestyle='dashed', label='+1 SD')
plt.axvline(mean_100 - standard_deviation_100, color='red', linestyle='dashed', label='-1 SD')
plt.legend(loc='upper right')
```

```python
plt.boxplot(dane_100, vert=True, showmeans= True, meanline=True)

# Calculate Q1 and Q3
q1 = np.quantile(dane_100, 0.25)
q3 = np.quantile(dane_100, 0.75)

plt.text(
    x=1.1,          
    y=q1,
    s=f"Q1 = {q1:.2f}",
    verticalalignment='center',
    color='black'
)

plt.text(
    x=1.1,
    y=q3,
    s=f"Q3 = {q3:.2f}",
    verticalalignment='center',
    color='black'
)

plt.text(
    x=1.1,
    y=median_100,
    s=f"Medina = {median_100:.2f}",
    verticalalignment='center',
    color='orange'
)

plt.text(
    x=1.1,
    y=np.min(dane_100),
    s=f"Min = {np.min(dane_100):.2f}",
    verticalalignment='center',
    color='black'
)

plt.text(
    x=1.1,
    y=np.max(dane_100),
    s=f"Max = {np.max(dane_100):.2f}",
    verticalalignment='center',
    color='black'
)

plt.text(
    x=0.7,
    y=np.mean(dane_100),
    s=f"Srednia = {np.mean(dane_100):.2f}",
    verticalalignment='center',
    color='green'
)

plt.show()
```
